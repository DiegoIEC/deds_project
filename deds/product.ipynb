```python
import pyodbc
import time

DB = {
    'servername': 'tcp:dwhunitedoutdoors.database.windows.net,1433',
    'database': 'DWH_United_Outdoors',
    'username': 'onderzoeker',
    'password': 'Wachtwoord1.'
}

conn_str = (
    'DRIVER={ODBC Driver 17 for SQL Server};'
    'SERVER=' + DB['servername'] + ';'
    'DATABASE=' + DB['database'] + ';'
    'UID=' + DB['username'] + ';'
    'PWD=' + DB['password'] + ';'
    'Encrypt=yes;'
    'TrustServerCertificate=no;'
    'Connection Timeout=60;'
)

def execute_batch_insert(cursor, queries):
    try:
        for query in queries:
            cursor.execute(query)
        cursor.connection.commit()
    except pyodbc.Error as e:
        print("An error occurred:", e)
        cursor.connection.rollback()
        cursor.connection.close()
        cursor = reconnect()
    return cursor

def reconnect():
    conn = pyodbc.connect(conn_str)
    return conn.cursor()

# Create the initial connection and cursor
export_cursor = reconnect()
batch_size = 1000
batch_queries = []

# Assuming work_order_merge is your DataFrame
for index, row in work_order_merge.iterrows():
    query = f"""
    INSERT INTO WORKORDER 
    (WORKORDER_id, WORKORDER_product_id, WORKORDER_order_quantity, WORKORDER_stocked_quantity, 
    WORKORDER_scrapped_quantity, WORKORDER_start_date, WORKORDER_end_date, WORKORDER_due_date, 
    WORKORDER_scrap_reason_name, WORKORDER_operation_sequence, WORKORDER_location_id, 
    WORKORDER_scheduled_start_date, WORKORDER_scheduled_end_date, WORKORDER_actual_start_date, 
    WORKORDER_actual_end_date, WORKORDER_actual_resource_hours, WORKORDER_planned_cost, 
    WORKORDER_actual_cost, WORKORDER_modified_date) 
    VALUES 
    ({row['WorkOrderID']}, {row['ProductID']}, {row['OrderQty']}, {row['StockedQty']}, 
    {row['ScrappedQty']}, '{row['StartDate']}', '{row['EndDate']}', '{row['DueDate']}', 
    '{row['Name'].replace("'", "''")}', {row['OperationSequence']}, {row['LocationID']}, 
    '{row['ScheduledStartDate']}', '{row['ScheduledEndDate']}', '{row['ActualStartDate']}', 
    '{row['ActualEndDate']}', {row['ActualResourceHrs']}, {row['PlannedCost']}, 
    {row['ActualCost']}, '{row['ModifiedDate']}')
    """
    batch_queries.append(query)
    
    if len(batch_queries) == batch_size:
        export_cursor = execute_batch_insert(export_cursor, batch_queries)
        batch_queries = []

if batch_queries:
    export_cursor = execute_batch_insert(export_cursor, batch_queries)

export_cursor.close()
```

```python
def extract_dates(dataframes, date_columns):
    dates = []
    for df, columns in zip(dataframes, date_columns):
        for column in columns:
            if column in df.columns:
                dates.extend(df[column].dropna().tolist())
    return dates

dataframes = [product_merge, purchase_merge, productvendor_merge, product_inventory, production_location, work_order_merge, bill_of_materials_merge]
date_columns = [['ModifiedDate'], ['OrderDate', 'ShipDate', 'DueDate', 'ModifiedDate'],
                ['LastReceiptDate', 'ModifiedDate'], ['ModifiedDate'],
                ['ModifiedDate'], ['StartDate', 'EndDate', 'DueDate', 'ScheduledStartDate', 'ScheduledEndDate', 'ActualStartDate', 'ActualEndDate', 'ModifiedDate'],
                ['StartDate', 'EndDate', 'ModifiedDate']]

dates = extract_dates(dataframes, date_columns)

unique_dates = pd.to_datetime(list(set(dates)))

date_df = pd.DataFrame({
    'date': unique_dates,
    'day': unique_dates.day,
    'month': unique_dates.month,
    'year': unique_dates.year
})

date_df.sort_values(by='date', inplace=True)

date_df.reset_index(drop=True, inplace=True)

date_df
```

```python
export_cursor = export_conn.cursor()

for index, row in date_df.iterrows():
    try:
        insert_query = f"""
        INSERT INTO DATE (DATE_date, DATE_day, DATE_month, DATE_year) 
        VALUES ('{row['date']}', {row['day']}, {row['month']}, {row['year']})
        """
        export_cursor.execute(insert_query)
    except pyodbc.Error as e:
        print("An error occurred:", e)
        print(insert_query)

export_conn.commit()
export_cursor.close()
```

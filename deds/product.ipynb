```python
newQuery = "SELECT * FROM [Production].[Product]"
product = pd.read_sql(newQuery,DB_AW_conn)

newQuery = "SELECT * FROM [Production].[ProductModel]"
product_model = pd.read_sql(newQuery,DB_AW_conn)

newQuery = "SELECT * FROM [Production].[ProductCategory]"
product_category = pd.read_sql(newQuery,DB_AW_conn)

newQuery = "SELECT * FROM [Production].[ProductSubcategory]"
product_sub_category = pd.read_sql(newQuery,DB_AW_conn)

newQuery = "SELECT * FROM [Production].[ProductSubcategory]"
product_sub_category = pd.read_sql(newQuery,DB_AW_conn)

newQuery = "SELECT * FROM [Production].[Culture]"
product_culture = pd.read_sql(newQuery,DB_AW_conn)

newQuery = "SELECT * FROM [Production].[ProductModelProductDescriptionCulture]"
product_culture_description = pd.read_sql(newQuery,DB_AW_conn)
```

```python
product_merge = product.merge(product_model, left_on='ProductModelID', right_on='ProductModelID', how='left', suffixes=('', '_ProductModel'))
product_merge = product_merge.merge(product_sub_category, left_on='ProductSubcategoryID', right_on='ProductSubcategoryID', how='left', suffixes=('', '_ProductSubcategory'))
product_merge = product_merge.merge(product_category, left_on='ProductSubcategoryID', right_on='ProductCategoryID', how='left', suffixes=('', '_ProductCategory'))
product_merge = product_merge.merge(product_culture_description, left_on='ProductModelID', right_on='ProductModelID', how='left', suffixes=('', '_ProductCultureDescription'))
product_merge = product_merge.merge(product_culture, left_on='CultureID', right_on='CultureID', how='left', suffixes=('', '_ProductCulture'))
product_merge
```

```python
product_merge['Name_ProductModel'].fillna('', inplace=True)
product_merge['Name_ProductCulture'].fillna('', inplace=True)
product_merge['Name_ProductSubcategory'].fillna('', inplace=True)
product_merge['Name_ProductCategory'].fillna('', inplace=True)
product_merge['Color'].fillna('', inplace=True)

product_merge['StandardCost'].fillna(0, inplace=True)
product_merge['ListPrice'].fillna(0, inplace=True)
product_merge['ProductLine'].fillna('', inplace=True)
product_merge['Class'].fillna('', inplace=True)
product_merge['Style'].fillna('', inplace=True)
product_merge
```

```python
export_cursor = export_conn.cursor()  

for index, row in product_merge.iterrows():
    try:
        query = f"""
        INSERT INTO PRODUCT 
        (PRODUCT_id, PRODUCT_name, PRODUCT_standard_cost, PRODUCT_list_price, PRODUCT_LINE_code, 
        PRODUCT_CLASS_code, PRODUCT_STYLE_code, Product_MODEL_name, Product_CULTURE_name, 
        PRODUCT_SUBCATEGORY_name, PRODUCT_CATEGORY_name, PRODUCT_color, PRODUCT_modified_date) 
        VALUES 
        ({row['ProductID']}, '{row['Name'].replace("'", "''")}', {row['StandardCost']}, {row['ListPrice']}, 
        '{row['ProductLine']}', '{row['Class']}', '{row['Style']}', 
        '{row['Name_ProductModel'].replace("'", "''")}', 
        '{row['Name_ProductCulture'].replace("'", "''")}', 
        '{row['Name_ProductSubcategory'].replace("'", "''")}', 
        '{row['Name_ProductCategory'].replace("'", "''")}',  
        '{row['Color']}', '{row['ModifiedDate']}')
        """
        export_cursor.execute(query)
    except pyodbc.Error as e:
        print("An error occurred:", e)
        print(query)

export_conn.commit()  
export_cursor.close()  
```

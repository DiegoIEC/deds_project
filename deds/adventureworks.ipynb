```python
import pyodbc
import pandas as pd

# Function to insert data into Sales_Store
def insert_into_sales_store(Sales_Store, export_conn):
    try:
        # Create a cursor
        export_cursor = export_conn.cursor()

        # SQL to create the Sales_Store table
        create_table_query = """
        IF NOT EXISTS (SELECT * FROM sysobjects WHERE name='Sales_Store' AND xtype='U')
        CREATE TABLE Sales_Store (
            BusinessEntityID INT,
            Name NVARCHAR(255),
            SalesPersonID INT
        )
        """
        # Execute the table creation query
        export_cursor.execute(create_table_query)
        
        # Insert data into the Sales_Store table
        for index, row in Sales_Store.iterrows():
            try:
                query = f"""
                INSERT INTO Sales_Store (BusinessEntityID, Name, SalesPersonID)
                VALUES (
                    {row['BusinessEntityID'] if row['BusinessEntityID'] is not None else 'NULL'}, 
                    '{str(row['Name']).replace("'", "''") if row['Name'] is not None else 'NULL'}',
                    {row['SalesPersonID'] if row['SalesPersonID'] is not None else 'NULL'}
                )
                """
                export_cursor.execute(query)
            except pyodbc.Error as e:
                print("Error:", e)
                print("Query:", query)

        # Commit the transaction
        export_conn.commit()
    except pyodbc.Error as e:
        print("Error:", e)
    finally:
        # Close the cursor
        export_cursor.close()

# Example usage
# Assuming you have a DataFrame 'Sales_Store' and a connection 'export_conn'
# Sales_Store = pd.read_sql_query("SELECT * FROM Sales.Store", import_conn)
insert_into_sales_store(Sales_Store, export_conn)
```

# Connection met AdventureWorks maken

```python
import pyodbc
import pandas as pd

server = 'XBLAPTOP\SQLEXPRESS'
database = 'AdventureWorks2019'
driver = 'ODBC Driver 17 for SQL Server'

connection_string = (
    f'DRIVER={{{driver}}};'
    f'SERVER={server};'
    f'DATABASE={database};'
    'Trusted_Connection=yes;'
)


import_conn = pyodbc.connect(connection_string)

query = "SELECT * FROM HumanResources.Department" 

df = pd.read_sql(query, import_conn)

print(df)
```

## HumanResources tabellen naar DataFrame

```python
HumanResources_Department = pd.read_sql_query("SELECT * FROM HumanResources.Department", import_conn)
HumanResources_Shift = pd.read_sql_query("SELECT * FROM HumanResources.Shift", import_conn)

# HumanResources_Employee = pd.read_sql_query("SELECT * FROM HumanResources.Employee", import_conn) Werkt niet ivm unsupported datatype

query_employee = """
SELECT BusinessEntityID, NationalIDNumber, LoginID, JobTitle, CAST(BirthDate AS DATETIME) AS BirthDate, 
       MaritalStatus, Gender, HireDate, SalariedFlag, VacationHours, SickLeaveHours, CurrentFlag, 
       CAST(ModifiedDate AS DATETIME) AS ModifiedDate
FROM HumanResources.Employee
"""

HumanResources_Employee = pd.read_sql_query(query_employee, import_conn)
```

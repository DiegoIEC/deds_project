import PRODUCT INTO DATAWAREHOUSE

```python


def insert_into_combined_product(combined_product_df, export_conn):
    try:
        export_cursor = export_conn.cursor()

        # Create the Combined_Product table if it does not exist
        create_table_query = """
        IF NOT EXISTS (SELECT * FROM sysobjects WHERE name='Combined_Product' AND xtype='U')
        CREATE TABLE Combined_Product (
            ProductID INT,
            Name NVARCHAR(255),
            ProductNumber NVARCHAR(25),
            MakeFlag BIT,
            FinishedGoodsFlag BIT,
            Color NVARCHAR(15),
            SafetyStockLevel SMALLINT,
            ReorderPoint SMALLINT,
            StandardCost DECIMAL(19, 4),
            ListPrice DECIMAL(19, 4),
            ModifiedDate DATETIME,
            ProductName NVARCHAR(255),
            SupplierID INT,
            CategoryID INT,
            QuantityPerUnit NVARCHAR(50),
            UnitPrice DECIMAL(19, 4),
            UnitsInStock SMALLINT,
            UnitsOnOrder SMALLINT,
            ReorderLevel SMALLINT,
            Discontinued BIT
        )
        """
        export_cursor.execute(create_table_query)

        # Insert rows into the Combined_Product table
        for index, row in combined_product_df.iterrows():
            try:
                query = f"""
                INSERT INTO Combined_Product (
                    ProductID, Name, ProductNumber, MakeFlag, FinishedGoodsFlag, Color,
                    SafetyStockLevel, ReorderPoint, StandardCost, ListPrice, ModifiedDate,
                    ProductName, SupplierID, CategoryID, QuantityPerUnit, UnitPrice,
                    UnitsInStock, UnitsOnOrder, ReorderLevel, Discontinued
                ) VALUES (
                    {row['ProductID'] if row['ProductID'] is not None else 'NULL'},
                    '{str(row['Name']).replace("'", "''") if row['Name'] is not None else 'NULL'}',
                    '{str(row['ProductNumber']).replace("'", "''") if row['ProductNumber'] is not None else 'NULL'}',
                    {1 if row['MakeFlag'] else 0},
                    {1 if row['FinishedGoodsFlag'] else 0},
                    '{str(row['Color']).replace("'", "''") if row['Color'] is not None else 'NULL'}',
                    {row['SafetyStockLevel'] if row['SafetyStockLevel'] is not None else 'NULL'},
                    {row['ReorderPoint'] if row['ReorderPoint'] is not None else 'NULL'},
                    {row['StandardCost'] if row['StandardCost'] is not None else 'NULL'},
                    {row['ListPrice'] if row['ListPrice'] is not None else 'NULL'},
                    '{row['ModifiedDate']}' if row['ModifiedDate'] is not None else 'NULL',
                    '{str(row['ProductName']).replace("'", "''") if row['ProductName'] is not None else 'NULL'}',
                    {row['SupplierID'] if row['SupplierID'] is not None else 'NULL'},
                    {row['CategoryID'] if row['CategoryID'] is not None else 'NULL'},
                    '{str(row['QuantityPerUnit']).replace("'", "''") if row['QuantityPerUnit'] is not None else 'NULL'}',
                    {row['UnitPrice'] if row['UnitPrice'] is not None else 'NULL'},
                    {row['UnitsInStock'] if row['UnitsInStock'] is not None else 'NULL'},
                    {row['UnitsOnOrder'] if row['UnitsOnOrder'] is not None else 'NULL'},
                    {row['ReorderLevel'] if row['ReorderLevel'] is not None else 'NULL'},
                    {1 if row['Discontinued'] else 0}
                )
                """
                export_cursor.execute(query)
            except pyodbc.Error as e:
                print("Error:", e)
                print("Query:", query)

        export_conn.commit()
    except pyodbc.Error as e:
        print("Error:", e)
    finally:
        export_cursor.close()

```

```python

def insert_into_employees_territories(merged_df, export_conn):
    try:
        export_cursor = export_conn.cursor()

        create_table_query = """
        IF NOT EXISTS (SELECT * FROM sysobjects WHERE name='Employees_Territories' AND xtype='U')
        CREATE TABLE Employees_Territories (
            EmployeeID INT,
            LastName NVARCHAR(255),
            FirstName NVARCHAR(255),
            Title NVARCHAR(255),
            TitleOfCourtesy NVARCHAR(255),
            BirthDate DATETIME,
            HireDate DATETIME,
            Address NVARCHAR(255),
            City NVARCHAR(255),
            Region NVARCHAR(255),
            PostalCode NVARCHAR(20),
            Country NVARCHAR(50),
            HomePhone NVARCHAR(25),
            Extension NVARCHAR(4),
            Photo IMAGE,
            Notes NVARCHAR(MAX),
            ReportsTo INT,
            PhotoPath NVARCHAR(255),
            TerritoryID NVARCHAR(20),
            TerritoryDescription NVARCHAR(255),
            RegionID INT
        )
        """
        export_cursor.execute(create_table_query)
        
        for index, row in merged_df.iterrows():
            try:
                query = f"""
                INSERT INTO Employees_Territories (EmployeeID, LastName, FirstName, Title, TitleOfCourtesy, BirthDate, HireDate, Address, City, Region, PostalCode, Country, HomePhone, Extension, Photo, Notes, ReportsTo, PhotoPath, TerritoryID, TerritoryDescription, RegionID)
                VALUES (
                    {row['EmployeeID'] if row['EmployeeID'] is not None else 'NULL'}, 
                    '{str(row['LastName']).replace("'", "''") if row['LastName'] is not None else 'NULL'}',
                    '{str(row['FirstName']).replace("'", "''") if row['FirstName'] is not None else 'NULL'}',
                    '{str(row['Title']).replace("'", "''") if row['Title'] is not None else 'NULL'}',
                    '{str(row['TitleOfCourtesy']).replace("'", "''") if row['TitleOfCourtesy'] is not None else 'NULL'}',
                    '{row['BirthDate']}' if row['BirthDate'] is not None else 'NULL',
                    '{row['HireDate']}' if row['HireDate'] is not None else 'NULL',
                    '{str(row['Address']).replace("'", "''") if row['Address'] is not None else 'NULL'}',
                    '{str(row['City']).replace("'", "''") if row['City'] is not None else 'NULL'}',
                    '{str(row['Region']).replace("'", "''") if row['Region'] is not None else 'NULL'}',
                    '{str(row['PostalCode']).replace("'", "''") if row['PostalCode'] is not None else 'NULL'}',
                    '{str(row['Country']).replace("'", "''") if row['Country'] is not None else 'NULL'}',
                    '{str(row['HomePhone']).replace("'", "''") if row['HomePhone'] is not None else 'NULL'}',
                    '{str(row['Extension']).replace("'", "''") if row['Extension'] is not None else 'NULL'}',
                    {row['Photo'] if row['Photo'] is not None else 'NULL'},
                    '{str(row['Notes']).replace("'", "''") if row['Notes'] is not None else 'NULL'}',
                    {row['ReportsTo'] if row['ReportsTo'] is not None else 'NULL'},
                    '{str(row['PhotoPath']).replace("'", "''") if row['PhotoPath'] is not None else 'NULL'}',
                    '{str(row['TerritoryID']).replace("'", "''") if row['TerritoryID'] is not None else 'NULL'}',
                    '{str(row['TerritoryDescription']).replace("'", "''") if row['TerritoryDescription'] is not None else 'NULL'}',
                    {row['RegionID'] if row['RegionID'] is not None else 'NULL'}
                )
                """
                export_cursor.execute(query)
            except pyodbc.Error as e:
                print("Error:", e)
                print("Query:", query)

        export_conn.commit()
    except pyodbc.Error as e:
        print("Error:", e)
    finally:
        export_cursor.close()

# Insert the merged DataFrame into the data warehouse
insert_into_employees_territories(merged_df, export_conn)

# Close the export connection
export_conn.close()
```

## Shortlist van relevante Afhankelijke Variabelen van AdventureWorks 

  1. SalesYTD (SalesPerson) 

   - Relevantie: 

     - Beslissingen over bonusstructuren en prestatiebeloningen. 

     - Planning van verkoopdoelstellingen en -strategieën voor komende periodes. 

   - Kenmerken: 

     - Geografische regio's (TerritoryID). 

     - Historische verkoopgegevens (SalesLastYear). 

  

2. TotalDue (SalesOrderHeader) 

   - Relevantie: 

     - Cashflow planning en financiële projecties. 

     - Beslissingen over kredietverlening aan klanten. 

   - Kenmerken: 

     - Besteldata (OrderDate, DueDate). 

     - Klantinformatie (CustomerID). 

  

3. UnitPrice (SalesOrderDetail) 

   - Relevantie: 

     - Prijszetting strategieën en promoties. 

     - Kostenbeheer en winstoptimalisatie. 

   - Kenmerken: 

     - Productinformatie (ProductID, SpecialOfferID). 

     - Verkoopkanaal (SalesPersonID). 

  

4. SalesQuota (SalesPerson) 

   - Relevantie: 

     - Evaluatie van verkoopdoelstellingen en prestatiebeoordelingen. 

     - Motivatie en beloningssystemen voor verkoopmedewerkers. 

   - Kenmerken: 

     - Verkoopprestaties (SalesYTD, SalesLastYear). 

     - Toegewezen regio's (TerritoryID). 

  

Bijpassende Onafhankelijke Variabelen 

  

Voor SalesYTD: 

   - TerritoryID: Geeft inzicht in de geografische locatie en kan verband houden met regionale verkooptrends. 

   - SalesLastYear: Historische verkoopgegevens om trends en seizoensinvloeden te identificeren. 

  

Voor TotalDue: 

   - OrderDate en DueDate: Tijdstippen van bestellingen en betalingen, belangrijk voor het begrijpen van cashflow cycli. 

   - CustomerID: Identificeert de klant en hun aankoopgeschiedenis, wat nuttig kan zijn voor het voorspellen van toekomstige uitgaven. 

  

Voor UnitPrice: 

   - ProductID en SpecialOfferID: Product- en promotieinformatie die invloed hebben op de verkoopprijs. 

   - SalesPersonID: Verkoopkanaal informatie, wat kan helpen bij het analyseren van prijsstrategieën en hun effectiviteit. 

  

Voor SalesQuota: 

   - SalesYTD en SalesLastYear: Huidige en historische verkoopprestaties die de haalbaarheid van verkoopquota kunnen beïnvloeden. 

   - TerritoryID: Geografische informatie die relevant kan zijn voor het vaststellen van realistische verkoopdoelen. 

  

Deze variabelen en kenmerken vormen een goede basis voor het opzetten van voorspellende modellen die kunnen helpen bij strategische besluitvorming binnen de organisatie.
